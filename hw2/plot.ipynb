```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = './data/train.csv'

with open(a, 'r') as f:
    lines = f.readlines()
    r = []
    for line in lines[1:]:
        b, c = line.split(',')
        r.append([int(x) for x in c.split()])
    d = max([max(x) for x in r])
    print(d)
```

```python
BPR_public = [0.04933, 0.05139, 0.05201, 0.05485]
BPR_private = [0.04404, 0.04787, 0.04800, 0.05098]

BCE_public = [0.04694, 0.05021, 0.04986, 0.05418]
BCE_private = [0.04145, 0.04417, 0.04641, 0.04771]

d = [16, 32, 64, 128]

fig, ax = plt.subplots(1, 2, figsize=(14,4))
ax[0].plot(d, BPR_public, '-o', label='public')
ax[0].plot(d, BPR_private, '-o', label='private')
ax[0].set_title('BPR',fontsize=18)
ax[0].set_xlabel('hidden dim.',fontfamily = 'sans-serif',fontstyle='italic')
ax[0].set_ylabel('MAP Score', fontfamily = 'sans-serif', fontstyle='oblique')
ax[0].set_xticks(d)
ax[0].legend()

ax[1].plot(d, BCE_public, '-o', label='public')
ax[1].plot(d, BCE_private, '-o', label='private')
ax[1].set_title('BCE',fontsize=18)
ax[1].set_xlabel('hidden dim.',fontfamily = 'sans-serif',fontstyle='italic')
ax[1].set_ylabel('MAP Score', fontfamily = 'sans-serif', fontstyle='oblique')
ax[1].set_xticks(d)
ax[1].legend()
plt.savefig('./plot/BPR')
plt.show()
```
